```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
'''FIRST ITERATION OF ENERGY MODEL'''
```

```python
''' Daily EV energy demand
Assumptions : 1000 car spaces, 10% EV spaces - 100 EV spaces
50% slow charging (3kWh), 50% fast charging (7kWh)
Slow charging takes 8 hours to complete, fast charging takes 4 hours to complete'''

t = np.arange(0,25) # hours in a day
x1 = [0,0,0,0,0,0,0,65,155,260,270,299.5,281.5,248,201.5,169,110.5,51,37.5,25.5,12,0,0,0,0] # kWh data from excel calculations

plt.plot(t,x1)
plt.title("Daily EV energy demand")
plt.xlabel("Time of day (hours)")
plt.ylabel("Energy (kWh)")
```

```python
# using txt file data
x1,y1 = np.loadtxt('EV_data.txt',delimiter = ',',unpack = True)
plt.plot(x1,y1)
plt.title("Daily EV energy demand")
plt.xlabel("Time of day (hours)")
plt.ylabel("Energy (kWh)")
```
